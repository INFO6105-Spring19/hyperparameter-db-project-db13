```python
import pandas as pd

lst = list(range(1,62))
lst1 = []
lst2 = []
for x in lst:
    lst1 = pd.read_json("CAb9R3kai_GBM_grid_1_AutoML_20190419_181340_model_"+str(x)+"_hy_parameter.json") 
    lst2.append(lst1)
print(lst2)
```

```python
temp = list()
for x in list(range(0,61)):
    df = pd.DataFrame(lst2[x])
    df = df.drop('default')
    df = df.drop('training_frame', axis=1)
    temp.append(df)
```

```python
final = pd.DataFrame()
for x in list(range(0,61)):
    x = pd.DataFrame(temp[x])
    final = pd.concat([final,x],axis =0)
```

```python
final.to_csv('CAb9R3kai_GBM_grid_1_AutoML_20190419_181340_model_.csv')
```
